## What age group is likely to get Heart Attack?

0 - No heart Attack
1 - heart Attack

```python
import pandas as pd
#matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
```

## Understanding and filtering Data

```python
df.head()
```

```python
df.describe()
```

```python
df.output.unique()
```

Sorting Data based on output = 1 and eliminating all rows which has output = 0, because we need only age whoch has chances of getting heart attack

```python
temp_df = df.groupby('output')
heart_attach_df = temp_df.get_group(1)
heart_attach_df
```

```python
heart_attach_df['range'] = pd.cut(heart_attach_df['age'], [20, 30, 40, 50, 60, 70, 80])
heart_attach_df
```

## Grouping Data 

```python
final_df = heart_attach_df.groupby('range')
final_dff = pd.DataFrame(final_df.count())
final_dff['percentage'] = final_dff['output']/final_dff['output'].sum()*100
final_dff = final_dff.reset_index()
final_dff['age_group'] = [20, 30, 40, 50, 60, 70]
final_dff
```

## Ploting the Bar Graph agains Age_Group and Percentage of getting Heart Attack

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(final_dff['age_group'], final_dff['percentage'], color ='maroon',
        width = 0.4)
plt.xlabel("Age Groups")
plt.ylabel("Percentage of getting heart Attack")
plt.title("Heart Attack for Different Age Group")
plt.show()
```
